```python
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic = pd.read_csv('titanic_data.csv')
```

```python
titanic.drop('PassengerId', axis=1, inplace=True)
```

```python
#alterando o nome das colunas
titanic.columns = ['Sobreviveu?', 'Classe', 'Nome', 'Sexo', 'Idade', 'Irmãos/Cônjugue', 'Pais/Filhos', 'Ticket', 'Tarifa', 'Cabine', 'Embarque']
```

```python
#Zerando valores nulos

titanic['Cabine'].fillna(0, inplace = True)
titanic['Idade'].fillna(0, inplace = True)

#Alterando valores das colunas
titanic['Sobreviveu?'].replace([0, 1], ['Sim', 'Não'], inplace = True)
titanic['Sexo'].replace(['male', 'female'], ['Masculino', 'Feminino'], inplace = True)
titanic['Classe'].replace([3, 1, 2], ['A', 'B', 'C'], inplace = True)
#titanic.sort_values('Nome', ascending=True, inplace = True)
```

```python
passageiros = sns.countplot(data=titanic, x='Sobreviveu?', color = 'green')
plt.title('Número de sobreviventes por sexo\n Total de passageiros: %d' % (titanic.shape[0]))
plt.xlabel('Sobreviventes')
plt.ylabel('Número de sobreviventes')
plt.show()
```

```python
sobreviventes_c = titanic.copy()
sobreviventes_c['Sobreviveu?'] = titanic['Sobreviveu?'] == 'Sim'
classe = sns.barplot(data=sobreviventes_c, y='Sobreviveu?', x='Classe')
plt.xlabel('Classe')
plt.ylabel('Proporção')
plt.show()
```

```python
qtd_sobreviventes = titanic[titanic['Sobreviveu?'] == 'Sim']['Classe'].value_counts().sort_values(ascending = True)
qtd_sobreviventes.plot(kind='barh', figsize = (11, 7), color='red')
plt.title('Quantidade de sobreviventes por classe\n Total: %d' % (qtd_sobreviventes.sum()))
plt.xlabel('Número de sobreviventes')
plt.ylabel('Classes')
plt.show()
```
