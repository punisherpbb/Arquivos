```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
plt.style.use('ggplot')
```

```python
#Lendo arquivo csv
data = pd.read_csv('vgsales.csv')
#Numero de linhas e colunas
print("Numero de linha: %d Numero de colunas %d" % (data.shape[0], data.shape[1]))
```

```python
#Renomeando colunas
data.columns = ['Rank', 'Nome', 'Plataforma', 'Ano', 'Genero', 'Editora', 'Vendas America do Norte', 'Vendas Estados Unidos', 'Vendas Jap√£o', 'Outras vendas', 'Vendas globais']
data.head()
```

```python
#Valores nulos
#data[data['Ano'].isnull()].head()
```

```python
#Jogos criados por plataforma(grafico)
valores_plataforma = data['Plataforma'].value_counts().head(10).plot(kind='bar', grid = False, figsize = (11, 5), rot = 0, color = 'green')

#Defini√ß√µes
plt.title('Jogos por plataforma')
plt.xlabel('Plataforma')
plt.ylabel('N√∫mero de jogos lan√ßados')
plt.show()
```

```python
#Quantidade de vendas no mundo
top_10 = data[['Nome', 'Vendas globais']].head(10).set_index('Nome').sort_values('Vendas globais', ascending = True)
top_10.plot(kind='barh', figsize = (11, 7), grid = False, color = 'darkred', legend = True)

plt.title('Jogos mais vendidos no mundo')
plt.xlabel('Vendas(em milh√µes de dolares)')
plt.show()
```

```python
top10_edit = data['Editora'].value_counts().head(10).sort_values(ascending = True)
top10_final = top10_edit.plot(kind='barh', figsize = (11, 7), rot = 0, grid=False, color='red')
plt.title('N√∫mero de jogos publicados por editora')
plt.xlabel('Numero de jogos publicados')
plt.ylabel('Editora')
plt.show()
```
